```python
print("Acessando o elemento da lista")
s = [1,2,4,67,86,4,3,6,8]
s[3]
```

```python
s[1:3] # Listas permitem acessar fatias (slices) dos seus valores, para obter a fatia da lista s compreendida entre os índices [1, 3)
```

```python
s[1:7:2] # Também é possível definir o intervalo da fatia com início, fim e incremento. O código abaixo seleciona os índices {1, 3 e 5}
```

```python
s[:3] # Se o início não for fornecido, a fatia irá iniciar do índice 0. 
```

```python
s[2:] # Se o fim não for definido, a fatia irá até a última posição da lista.
```

```python
print("modificando a lista")
s[2] = 5 # troca o valor do index 2(4) para 5
s
```

```python
s[1:3] = [3,8] # Uma fatia inteira pode ser substituída pelos valores de uma outra lista do mesmo tamanho
s
```

```python
del s[1:7:2] # O operador del permite remover uma fatia inteira da lista, reduzindo seu tamanho
s
```

```python
s.append(0) # adiciona um novo valor ao final da lista
s
```

```python
s.insert(0,3) # adiciona elemento na posição desejada
s
```

```python
s.extend([2,5]) # adiciona elementos de outra lista ao final da lista
s
```

```python
print("formatando strings")
s = 'Média: 1, Variância: 4'
s
```

```python
print("outras operações com string")
'Média: 1, Variância: 4'.split(sep=', ') # split -> divide a string em uma lista contendo substrings delimitadas pela string sep. Caso sep não seja informada, espaços em branco consecutivos serão tratados como delimitadores.
```

```python
'Média: 1, Variância: 4'.replace(': ', '=') # replace -> ubstitui todas as ocorrências de uma substring por uma nova substring e dá como resultado a string modificada.
```

```python
'Média: 1, Variância: 4'.replace(': ', '=', 1) #  Opcionalmente, pode-se informar quantas substituições deseja-se fazer.
```

```python
'Estatística'.find('t') # find -> permite encontrar o índice da primeira ocorrência de uma substring em uma string ou em uma fatia de uma string.
```

```python
'Estatística'.find('t', 3, 7)
```

```python
'CIÊNCIA DE DADOS'.lower() # coloca todos os caracteres em caixa baixa (minúsculas)
```

```python
'ciência de dados'.upper() #  coloca todos os caracteres em caixa alta (maiúsculas)
```

```python
'ciência de dados'.title() # coloca a string em formato de título,as primeiras letras das palavras ficam maiúsculas e as demais minúsculas (note que o algoritmo faz essa transformação em todas as palavras da string)
```
