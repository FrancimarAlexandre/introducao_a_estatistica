```python
print("Acessando o elemento da lista")
s = [1,2,4,67,86,4,3,6,8]
s[3]
```

```python
s[1:3] # Listas permitem acessar fatias (slices) dos seus valores, para obter a fatia da lista s compreendida entre os índices [1, 3)
```

```python
s[1:7:2] # Também é possível definir o intervalo da fatia com início, fim e incremento. O código abaixo seleciona os índices {1, 3 e 5}
```

```python
s[:3] # Se o início não for fornecido, a fatia irá iniciar do índice 0. 
```

```python
s[2:] # Se o fim não for definido, a fatia irá até a última posição da lista.
```

```python
print("modificando a lista")
s[2] = 5 # troca o valor do index 2(4) para 5
s
```

```python
s[1:3] = [3,8] # Uma fatia inteira pode ser substituída pelos valores de uma outra lista do mesmo tamanho
s
```

```python
del s[1:7:2] # O operador del permite remover uma fatia inteira da lista, reduzindo seu tamanho
s
```

```python
s.append(0) # adiciona um novo valor ao final da lista
s
```

```python
s.insert(0,3) # adiciona elemento na posição desejada
s
```

```python
s.extend([2,5]) # adiciona elementos de outra lista ao final da lista
s
```

```python
print("Comparações de conjuntos")
S1 = {1, 2}
S2 = {1, 2, 3, 4}

S1 <= S2
```

```python
S1 < S2
```

```python
S1 >= S2
```

```python
S1 > S2
```

```python
#A comparação de igualdade de dois conjuntos só retorna True sse S1⊆S2 e S1⊇S2, i.e. se ambos os conjuntos possuem exatamente os mesmos elementos:
{1, 2} == {1, 2}
```

```python
S1 = {0, 8}
S2 = {1, 2, 3, 4}

S1.isdisjoint(S2) # Também pode ser útil verificar se dois conjuntos são disjuntos:
```

```python
print("Operações com dois ou mais conjuntos")
# Python também oferece operadores para realizar a união (|), interseção (&) e a diferença (-) de dois ou mais conjuntos e a diferença simétrica (^) entre dois conjuntos.
S1 = {1, 2, 3}
S2 = {2, 3, 4}
S3 = {5, 3, 4}

S1 | S2 | S3 
```

```python
S1 & S2 & S3
```

```python
S1 - S2 - S3
```

```python
S1 ^ S2
```
